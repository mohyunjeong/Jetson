```python
import evdev

print(evdev.list_devices())

path = '/dev/input/event2'

device = evdev.InputDevice(path)
device
```

```python
import custom
import head

from evdev import ecodes, categorize

for event in device.read_loop():
    
    if event.type == ecodes.EV_KEY: # ë””ì§€í„¸ ì‹ í˜¸
        
        if event.value == 1: # ë²„íŠ¼ 1 or 0 ì‹ í˜¸
            
            print(event)
            
            if event.code == 288 : 
                print("ğŸ’«1 ë²„íŠ¼ í´ë¦­ğŸ’«") 
            elif event.code == 289 : 
                print("ğŸ’«2 ë²„íŠ¼ í´ë¦­ğŸ’«")                
            elif event.code == 290 :   
                print("ğŸ’«3 ë²„íŠ¼ í´ë¦­ğŸ’«")  
            elif event.code == 291 : 
                print("ğŸ’«4 ë²„íŠ¼ í´ë¦­ğŸ’«")        
            elif event.code == 292 :   
                print("ğŸ’«L1 ë²„íŠ¼ í´ë¦­ğŸ’«")    
            elif event.code == 293 :
                print("ğŸ’«R1 ë²„íŠ¼ í´ë¦­ğŸ’«")      
            elif event.code == 294 :
                print("ğŸ’«L2 ë²„íŠ¼ í´ë¦­ğŸ’«")     
            elif event.code == 295 :
                print("ğŸ’«R2 ë²„íŠ¼ í´ë¦­ğŸ’«")                    
            elif event.code == 296 :
                print("ğŸ’«Select ë²„íŠ¼ í´ë¦­ğŸ’«")
                head.center()
            elif event.code == 297 :
                print("ğŸ’«Start ë²„íŠ¼ í´ë¦­ğŸ’«")
                custom.stop()
            elif event.code == 298 :
                print("ğŸ’«ì™¼ìª½ ì¡°ì´ìŠ¤í‹± í´ë¦­ğŸ’«")    
            elif event.code == 299 :
                print("ğŸ’«ì˜¤ë¥¸ìª½ ì¡°ì´ìŠ¤í‹± í´ë¦­ğŸ’«")
                
    elif event.type == ecodes.EV_ABS : # ì•„ë‚ ë¡œê·¸ ì‹ í˜¸
        
        absolute = categorize(event) # import categorize
        code = ecodes.bytype[absolute.event.type][absolute.event.code]
        value = absolute.event.value
        
        # print(code) # ABS_Z, ABS_X, ABS_RX, ABS_RZ, ABS_Y
        
        if code == "ABS_Y" :
            if value == 0 :
                print("ğŸ’«ì™¼ìª½ ì¡°ì´ìŠ¤í‹± UPğŸ’«")
                head.tilt_up()
            elif value == 255 :
                print("ğŸ’«ì™¼ìª½ ì¡°ì´ìŠ¤í‹± DOWNğŸ’«")
                head.tilt_down()
        elif code == "ABS_X" :
            if value == 0 :
                print("ğŸ’«ì™¼ìª½ ì¡°ì´ìŠ¤í‹± LEFTğŸ’«")
                head.pan_left()
            if value == 255 :
                print("ğŸ’«ì™¼ìª½ ì¡°ì´ìŠ¤í‹± RIGHTğŸ’«")
                head.pan_right()
        elif code == "ABS_RZ" :
            if value == 0 :
                print("ğŸ’«ì˜¤ë¥¸ìª½ ì¡°ì´ìŠ¤í‹± UPğŸ’«")
                custom.forward()
            elif value == 255 :
                print("ğŸ’«ì˜¤ë¥¸ìª½ ì¡°ì´ìŠ¤í‹± DOWNğŸ’«")
                custom.backward()
        elif code == "ABS_RX" :
            if value == 0 :
                print("ğŸ’«ì˜¤ë¥¸ìª½ ì¡°ì´ìŠ¤í‹± LEFTğŸ’«")
                custom.left()
            elif value == 255 :
                print("ğŸ’«ì˜¤ë¥¸ìª½ ì¡°ì´ìŠ¤í‹± RIGHTğŸ’«")
                custom.right()
```
